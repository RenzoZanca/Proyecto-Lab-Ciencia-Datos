#### üìå Batch 1 (01/01/25 - 05/01/25) [0.75 puntos]

````python
import pandas as pd

def prepare_data(batch_path, prediction_path):
    batch = pd.read_parquet(f'data/{batch_path}.parquet')
    pred = pd.read_csv(f'predictions/{prediction_path}.csv')

    # process predictions
    first_row = pred.columns.tolist()
    pred.columns = ['customer_id', 'product_id']
    pred.loc[-1] = first_row  
    pred.index = pred.index + 1
    pred = pred.sort_index()
    print(f"Number of predictions: {len(pred)}")
    
    # process batch
    combinations = batch[['customer_id', 'product_id']].drop_duplicates()
    print(f"Number of combinations: {len(combinations)}")

    return pred, combinations

    

pred1, combinations1 = prepare_data('batch_t1', 'predictions1')
````

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_predictions(predictions, ground_truth):
    # Asegurar columnas correctas
    predictions.columns = ['customer_id', 'product_id']
    ground_truth.columns = ['customer_id', 'product_id']

    pred = predictions.copy()

    # Agregar columna de etiqueta (1 si existe en ground_truth)
    pred['label'] = pred.apply(
        lambda row: 1 if ((ground_truth['customer_id'] == row['customer_id']) &
                          (ground_truth['product_id'] == row['product_id'])).any() else 0,
        axis=1
    )

    # Ground truth real: 1 si (customer_id, product_id) est√° presente
    y_true = pred['label']
    y_pred = [1] * len(pred)  # ya que todos en predictions son predichos como positivos

    # M√©tricas
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)

    print(f"Accuracy: {acc}, Precision: {prec}, Recall: {rec}, F1 Score: {f1}")

    return {'accuracy': acc, 'precision': prec, 'recall': rec, 'f1': f1}

metrics1 = evaluate_predictions(pred1, combinations1)
````

````python
from collections import Counter
import seaborn as sns
import matplotlib.pyplot as plt

def deep_batch_analysis(pred_df, gt_df, productos_df=None):
    # 1) Marcar verdadero/falso positivo
    merged = pred_df.merge(gt_df.assign(label=1), 
                           on=['customer_id', 'product_id'], 
                           how='left')
    merged['label'] = merged['label'].fillna(0)

    tp = int(merged['label'].sum())
    fp = int((merged['label'] == 0).sum())
    fn = len(gt_df) - tp
    tn = 0  # por construcci√≥n no se listan negativos predichos
    
    print(f"TP={tp}, FP={fp}, FN={fn}")
    print(f"Cobertura (Recall): {tp / (tp + fn):.4f}")
    print(f"Tasa de predicciones: {len(pred_df) / len(gt_df):.2f}x")

    # 2) Distribuci√≥n de precisi√≥n por cliente
    per_client = merged.groupby('customer_id')['label'].agg(['sum', 'count'])
    per_client['precision_cliente'] = per_client['sum'] / per_client['count']
    
    plt.figure(figsize=(7,4))
    sns.histplot(per_client['precision_cliente'], bins=20, kde=False)
    plt.title('Distribuci√≥n de precisi√≥n por cliente')
    plt.xlabel('Precisi√≥n por cliente')
    plt.ylabel('Frecuencia')
    plt.show()

    # 3) Productos recomendados vs comprados
    per_client['comprados'] = gt_df.groupby('customer_id').size()
    per_client['recomendados'] = per_client['count']
    sample_n = 50
    subset = per_client.sample(sample_n, random_state=42)
    ax = subset[['comprados', 'recomendados']].plot(kind='bar', figsize=(11,4))
    ax.set_title('Productos comprados vs recomendados (muestra de 50 clientes)')
    ax.set_xlabel('customer_id')
    ax.set_ylabel('N¬∫ productos')
    plt.tight_layout()
    plt.show()

    # 4) An√°lisis de categor√≠as
    if productos_df is not None:
        merged_cat = merged.merge(productos_df[['product_id','category']], on='product_id', how='left')
        cat_stats = merged_cat.groupby('category')['label'].agg(['sum','count'])
        cat_stats['precision'] = cat_stats['sum'] / cat_stats['count']
        display(cat_stats.sort_values('precision', ascending=False).head(10))
````

````python
deep_batch_analysis(pred1, combinations1)
````

### üìå An√°lisis comparativo [2.0 puntos]

````python
# Crear el DataFrame
metrics_list = [metrics1, metrics2, metrics3, metrics4]
df_metrics = pd.DataFrame(metrics_list, index=["1", "2", "3", "4"])

pred_objs = [pred1, pred2, pred3, pred4]
comb_objs = [combinations1, combinations2, combinations3, combinations4]
df_metrics["tasa_pred"] = [len(p) / len(c) for p, c in zip(pred_objs, comb_objs)]

# --- C√°lculo program√°tico del recall real ---------------------------------
def recall_real(pred_df, gt_df):
    """Cobertura real: TP / (TP + FN) considerando todo el ground-truth."""
    merged = pred_df.merge(gt_df.assign(label=1),
                           on=['customer_id', 'product_id'],
                           how='left')
    tp = merged['label'].fillna(0).sum()
    fn = len(gt_df) - tp
    return tp / (tp + fn)

recall_real_vals = [
    recall_real(pred1, combinations1),
    recall_real(pred2, combinations2),
    recall_real(pred3, combinations3),
    recall_real(pred4, combinations4)
]

# --- A√±adir al DataFrame existente ----------------------------------------
df_metrics["recall_real"] = recall_real_vals


semanas = df_metrics.index

# ----- A) Recall y Recall Real -----------------------------------------------------------
plt.figure(figsize=(6,3))
sns.lineplot(x=df_metrics.index, y=df_metrics["recall_real"],
             marker="o", label="Recall real")
sns.lineplot(x=df_metrics.index, y=df_metrics["recall"],
             marker="o", linestyle="--", label="Recall interno")
plt.ylim(0.8, 1.05)
plt.xlabel("Semana")
plt.ylabel("Recall")
plt.title("Comparaci√≥n de ambos recalls")
plt.grid(True)
plt.legend()
plt.show()
````

> **Interpretaci√≥n**  
> - El ‚Äúrecall interno‚Äù se calcula solo sobre los pares recomendados y vale siempre 1,0 porque dentro de ese subconjunto no existen falsos negativos. El ‚Äúrecall real‚Äù compara contra todas las compras efectivas y refleja la verdadera cobertura del sistema. Para la evaluaci√≥n de negocio se considera √∫nicamente el recall real.
> - El recall alcanza su punto m√°ximo (0,91) en la semana 2, cuando el modelo incorpora por primera vez un volumen mayor de datos hist√≥ricos despu√©s del reentrenamiento. A partir de la semana 3 la m√©trica desciende, lo que indica que el crecimiento del set de recomendaciones no basta para mantener la misma cobertura. La ca√≠da m√°s pronunciada ocurre en la semana 4 y coincide con una reducci√≥n significativa del n√∫mero de recomendaciones emitidas, signo de que el modelo sacrifica cobertura cuando disminuye el volumen.

Existen dos formas de calcular el recall que se han usado en el notebook y, por lo tanto, aparecen dos valores diferentes:

1. Recall ‚Äúinterno a las recomendaciones‚Äù (siempre 1,0)  
   ‚Äì Se calcula dentro de `evaluate_predictions`.  
   ‚Äì Solo considera las combinaciones cliente-producto que el modelo ya recomend√≥.  
   ‚Äì Como `y_pred` se fija en 1 para todas esas filas, no existen falsos negativos y el recall resulta 1,0 por construcci√≥n.  
   ‚Äì Este valor no refleja la cobertura real; es √∫til √∫nicamente para verificar que, dentro del subconjunto recomendado, el c√≥digo etiquet√≥ correctamente los verdaderos positivos.

2. Recall ‚Äúreal‚Äù (0,84‚Äì0,91)  
   ‚Äì Se calcula en `deep_batch_analysis`.  
   ‚Äì Compara las recomendaciones con **todo** el universo de compras reales de la semana.  
   ‚Äì Incluye los pares cliente-producto que ocurrieron pero no fueron recomendados; por ello incorpora falsos negativos y ofrece la medida de cobertura pertinente para el negocio.

Si bien se mostrar√°n ambas, debiesemos enfocarnos realmente en el Recall "real (0.84-0.91)" ya que:
- es la m√©trica de cobertura que responde a la pregunta central: ‚Äú¬øCu√°ntas compras efectivas logra captar el modelo?‚Äù.

**II. Escenario filtrado (3 200 primeras filas)**


````python
# --------- Construcci√≥n del DF filtrado (Top-N = 3200) --------------------
topN = 3200
def metrics_topN(pred, gt, N):
    m = evaluate_predictions(pred[:N], gt)            # accuracy, precision, recall, f1
    m['tasa_pred'] = N / len(gt)
    m['recall_real'] = recall_real(pred[:N], gt)
    return m

df_metrics_topN = pd.DataFrame([
    metrics_topN(pred1, combinations1, topN),
    metrics_topN(pred2, combinations2, topN),
    metrics_topN(pred3, combinations3, topN),
    metrics_topN(pred4, combinations4, topN)
], index=["1","2","3","4"])


# --------- Funci√≥n auxiliar para dibujar l√≠neas dobles --------------------
def plot_metric_compare(metric, ymin, ymax, ylabel):
    plt.figure(figsize=(6,3))
    sns.lineplot(x=df_metrics.index, y=df_metrics[metric],
                 marker="o", label=f"{metric} original")
    sns.lineplot(x=df_metrics_topN.index, y=df_metrics_topN[metric],
                 marker="o", linestyle="--", label=f"{metric} Top-N")
    plt.ylim(ymin, ymax)
    plt.xlabel("Semana")
    plt.ylabel(ylabel)
    plt.title(f"{ylabel} original vs Top-N")
    plt.grid(True)
    plt.legend()
    plt.show()

# --------- Gr√°ficos solicitados -------------------------------------------
plot_metric_compare("precision", 0.0, 0.55, "Precisi√≥n")
plot_metric_compare("recall_real", 0, 1.1, "Recall real")
plot_metric_compare("f1", 0.0, 0.75, "F1")
plot_metric_compare("accuracy", 0.0, 0.55, "Accuracy")
````
