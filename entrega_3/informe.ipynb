### üìå An√°lisis individual [3.0 puntos]

#### üìå Batch 1 (01/01/25 - 05/01/25) [0.75 puntos]

````python
import pandas as pd

def prepare_data(batch_path, prediction_path):
    batch = pd.read_parquet(f'data/{batch_path}.parquet')
    pred = pd.read_csv(f'predictions/{prediction_path}.csv')

    # process predictions
    first_row = pred.columns.tolist()
    pred.columns = ['customer_id', 'product_id']
    pred.loc[-1] = first_row  
    pred.index = pred.index + 1
    pred = pred.sort_index()
    print(f"Number of predictions: {len(pred)}")
    
    # process batch
    combinations = batch[['customer_id', 'product_id']].drop_duplicates()
    print(f"Number of combinations: {len(combinations)}")

    return pred, combinations

    

pred1, combinations1 = prepare_data('batch_t1', 'predictions1')
````

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_predictions(predictions, ground_truth):
    # Asegurar columnas correctas
    predictions.columns = ['customer_id', 'product_id']
    ground_truth.columns = ['customer_id', 'product_id']

    pred = predictions.copy()

    # Agregar columna de etiqueta (1 si existe en ground_truth)
    pred['label'] = pred.apply(
        lambda row: 1 if ((ground_truth['customer_id'] == row['customer_id']) &
                          (ground_truth['product_id'] == row['product_id'])).any() else 0,
        axis=1
    )

    # Ground truth real: 1 si (customer_id, product_id) est√° presente
    y_true = pred['label']
    y_pred = [1] * len(pred)  # ya que todos en predictions son predichos como positivos

    # M√©tricas
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)

    print(f"Accuracy: {acc}, Precision: {prec}, Recall: {rec}, F1 Score: {f1}")

    return {'accuracy': acc, 'precision': prec, 'recall': rec, 'f1': f1}

metrics1 = evaluate_predictions(pred1, combinations1)
````

Notamos un mal desempe√±o general del modelo, pero un recall perfecto. Esto se debe a el gran tama√±o de las predicciones. Para probar, medimos el desempe√±o utilizando menos datos como predicci√≥n.

````python
evaluate_predictions(pred1[:3200], combinations1)
````

Obtenemos un desempe√±o ligeramente mejor, pero el precision sigue estando en un valor bajo.

#### üìå Batch 2 (06/01/25 - 12/01/25) [0.75 puntos]

````python
pred2, combinations2 = prepare_data('batch_t2', 'predictions2')
metrics2 = evaluate_predictions(pred2, combinations2)
````

#### üìå Batch 03 (13/01/25 - 19/01/25) [0.75 puntos]

````python
pred3, combinations3 = prepare_data('batch_t3', 'predictions3')
metrics3 = evaluate_predictions(pred3, combinations3)
````

#### üìå Batch 04 (20/01/25 - 26/01/25) [0.75 puntos]

````python
pred4, combinations4, = prepare_data('batch_t4', 'predictions4')
metrics4 = evaluate_predictions(pred4, combinations4)
````

### üìå An√°lisis comparativo [2.0 puntos]

````python
import matplotlib.pyplot as plt

def compare_predictions(preds_results, metrics):
    plt.figure(figsize=(10, 6))

    for metric in metrics:
        values = [pred[metric] for pred in preds_results]
        plt.plot(range(len(preds_results)), values, marker='o', label=metric.capitalize())

    plt.xticks(range(len(preds_results)), [f'Pred {i+1}' for i in range(len(preds_results))])
    plt.xlabel('Predictions')
    plt.ylabel('Score')
    plt.title('Comparison of Metrics Across Predictions')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Ejemplo de uso
metrics = [metrics1, metrics2, metrics3, metrics4]
compare_predictions(metrics, ['precision', 'f1'])
````

````python
compare_predictions(metrics, ['accuracy'])
````

````python
compare_predictions(metrics, ['f1'])
````

````python
compare_predictions(metrics, ['recall'])
````
