#### 📌 Batch 1 (01/01/25 - 05/01/25) [0.75 puntos]

````python
import pandas as pd

def prepare_data(batch_path, prediction_path):
    batch = pd.read_parquet(f'data/{batch_path}.parquet')
    pred = pd.read_csv(f'predictions/{prediction_path}.csv')

    # process predictions
    first_row = pred.columns.tolist()
    pred.columns = ['customer_id', 'product_id']
    pred.loc[-1] = first_row  
    pred.index = pred.index + 1
    pred = pred.sort_index()
    print(f"Number of predictions: {len(pred)}")
    
    # process batch
    combinations = batch[['customer_id', 'product_id']].drop_duplicates()
    print(f"Number of combinations: {len(combinations)}")

    return pred, combinations

    

pred1, combinations1 = prepare_data('batch_t1', 'predictions1')
````

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_predictions(predictions, ground_truth):
    # Asegurar columnas correctas
    predictions.columns = ['customer_id', 'product_id']
    ground_truth.columns = ['customer_id', 'product_id']

    pred = predictions.copy()

    # Agregar columna de etiqueta (1 si existe en ground_truth)
    pred['label'] = pred.apply(
        lambda row: 1 if ((ground_truth['customer_id'] == row['customer_id']) &
                          (ground_truth['product_id'] == row['product_id'])).any() else 0,
        axis=1
    )

    # Ground truth real: 1 si (customer_id, product_id) está presente
    y_true = pred['label']
    y_pred = [1] * len(pred)  # ya que todos en predictions son predichos como positivos

    # Métricas
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)

    print(f"Accuracy: {acc}, Precision: {prec}, Recall: {rec}, F1 Score: {f1}")

    return {'accuracy': acc, 'precision': prec, 'recall': rec, 'f1': f1}

metrics1 = evaluate_predictions(pred1, combinations1)
````

````python
from collections import Counter
import seaborn as sns
import matplotlib.pyplot as plt

def deep_batch_analysis(pred_df, gt_df, productos_df=None):
    # 1) Marcar verdadero/falso positivo
    merged = pred_df.merge(gt_df.assign(label=1), 
                           on=['customer_id', 'product_id'], 
                           how='left')
    merged['label'] = merged['label'].fillna(0)

    tp = int(merged['label'].sum())
    fp = int((merged['label'] == 0).sum())
    fn = len(gt_df) - tp
    tn = 0  # por construcción no se listan negativos predichos
    
    print(f"TP={tp}, FP={fp}, FN={fn}")
    print(f"Cobertura (Recall): {tp / (tp + fn):.4f}")
    print(f"Tasa de predicciones: {len(pred_df) / len(gt_df):.2f}x")

    # 2) Distribución de precisión por cliente
    per_client = merged.groupby('customer_id')['label'].agg(['sum', 'count'])
    per_client['precision_cliente'] = per_client['sum'] / per_client['count']
    
    plt.figure(figsize=(7,4))
    sns.histplot(per_client['precision_cliente'], bins=20, kde=False)
    plt.title('Distribución de precisión por cliente')
    plt.xlabel('Precisión por cliente')
    plt.ylabel('Frecuencia')
    plt.show()

    # 3) Productos recomendados vs comprados
    per_client['comprados'] = gt_df.groupby('customer_id').size()
    per_client['recomendados'] = per_client['count']
    sample_n = 50
    subset = per_client.sample(sample_n, random_state=42)
    ax = subset[['comprados', 'recomendados']].plot(kind='bar', figsize=(11,4))
    ax.set_title('Productos comprados vs recomendados (muestra de 50 clientes)')
    ax.set_xlabel('customer_id')
    ax.set_ylabel('Nº productos')
    plt.tight_layout()
    plt.show()

    # 4) Análisis de categorías
    if productos_df is not None:
        merged_cat = merged.merge(productos_df[['product_id','category']], on='product_id', how='left')
        cat_stats = merged_cat.groupby('category')['label'].agg(['sum','count'])
        cat_stats['precision'] = cat_stats['sum'] / cat_stats['count']
        display(cat_stats.sort_values('precision', ascending=False).head(10))
````

````python
deep_batch_analysis(pred1, combinations1)
````

Notamos un mal desempeño general del modelo, pero un recall perfecto. Esto se debe a el gran tamaño de las predicciones. Para probar, medimos el desempeño utilizando menos datos como predicción.

De estas cifras se derivan dos indicadores clave:

   El sistema identifica 85 % de las compras efectivas, lo que garantiza que la gran mayoría de los productos realmente adquiridos queda recogida en la lista de recomendaciones.

2. Tasa de predicciones = |predicciones| / |combinaciones reales| = 31,8  
   Por cada interacción real cliente-producto el modelo emite casi 32 sugerencias, lo que explica la precisión global cercana a 0,03 reportada previamente.

#### Distribución de precisión por cliente  
El histograma evidencia que más de 60 % de los clientes presenta precisión igual a cero y que menos de 10 % supera 0,10. Este patrón confirma que el modelo discrimina de forma muy desigual: para una minoría de clientes acierta con cierta frecuencia, mientras que para la mayoría las recomendaciones son esencialmente aleatorias.

#### Productos comprados versus recomendados  
En una muestra de 50 clientes (gráfico resumido) el número de productos recomendados por cliente oscila entre 40 y 90, frente a un máximo de 15 productos realmente comprados. La brecha subraya que la política de “probabilidad > 0” es excesivamente permisiva y provoca una avalancha de sugerencias irrelevantes.

#### Sensibilidad a la reducción de recomendaciones  
Como ejercicio exploratorio se repitió la evaluación limitando la salida a las primeras 3 200 filas. La precisión aumentó de 0,03 a 0,07 y el F1 se duplicó, mientras que el recall se mantuvo en 1,0. Este resultado demuestra que un umbral de probabilidad más estricto —o un esquema top-N por cliente— permitiría mejorar sustancialmente el balance precisión–recall sin sacrificar cobertura.

#### Implicancias operativas  
• Con la configuración actual el coste comercial de emitir 32 recomendaciones por cada compra real resulta inasumible: la saturación de ofertas puede generar rechazo en los clientes.  
• El ajuste del umbral de decisión se perfila como la medida de mayor impacto inmediato; incluso un recorte del 90 % de las predicciones mantendría una cobertura elevada.  
• La heterogeneidad de precisión sugiere explorar un enfoque segmentado (por tipo de cliente o categoría de producto) o incorporar un filtro post-modelo que limite el número máximo de recomendaciones por cliente.

Estas conclusiones sirven de base para los ajustes propuestos y para el análisis comparativo entre semanas.

````python
# Crear el DataFrame
metrics_list = [metrics1, metrics2, metrics3, metrics4]
df_metrics = pd.DataFrame(metrics_list, index=["1", "2", "3", "4"])

pred_objs = [pred1, pred2, pred3, pred4]
comb_objs = [combinations1, combinations2, combinations3, combinations4]
df_metrics["tasa_pred"] = [len(p) / len(c) for p, c in zip(pred_objs, comb_objs)]

# --- Cálculo programático del recall real ---------------------------------
def recall_real(pred_df, gt_df):
    """Cobertura real: TP / (TP + FN) considerando todo el ground-truth."""
    merged = pred_df.merge(gt_df.assign(label=1),
                           on=['customer_id', 'product_id'],
                           how='left')
    tp = merged['label'].fillna(0).sum()
    fn = len(gt_df) - tp
    return tp / (tp + fn)

recall_real_vals = [
    recall_real(pred1, combinations1),
    recall_real(pred2, combinations2),
    recall_real(pred3, combinations3),
    recall_real(pred4, combinations4)
]

# --- Añadir al DataFrame existente ----------------------------------------
df_metrics["recall_real"] = recall_real_vals


semanas = df_metrics.index

# ----- A) Recall y Recall Real -----------------------------------------------------------
plt.figure(figsize=(6,3))
sns.lineplot(x=df_metrics.index, y=df_metrics["recall_real"],
             marker="o", label="Recall real")
sns.lineplot(x=df_metrics.index, y=df_metrics["recall"],
             marker="o", linestyle="--", label="Recall interno")
plt.ylim(0.8, 1.05)
plt.xlabel("Semana")
plt.ylabel("Recall")
plt.title("Comparación de ambos recalls")
plt.grid(True)
plt.legend()
plt.show()
````

> **Interpretación**  
> - El “recall interno” se calcula solo sobre los pares recomendados y vale siempre 1,0 porque dentro de ese subconjunto no existen falsos negativos. El “recall real” compara contra todas las compras efectivas y refleja la verdadera cobertura del sistema. Para la evaluación de negocio se considera únicamente el recall real.
> - El recall alcanza su punto máximo (0,91) en la semana 2, cuando el modelo incorpora por primera vez un volumen mayor de datos históricos después del reentrenamiento. A partir de la semana 3 la métrica desciende, lo que indica que el crecimiento del set de recomendaciones no basta para mantener la misma cobertura. La caída más pronunciada ocurre en la semana 4 y coincide con una reducción significativa del número de recomendaciones emitidas, signo de que el modelo sacrifica cobertura cuando disminuye el volumen.

Existen dos formas de calcular el recall que se han usado en el notebook y, por lo tanto, aparecen dos valores diferentes:

1. Recall “interno a las recomendaciones” (siempre 1,0)  
   – Se calcula dentro de `evaluate_predictions`.  
   – Solo considera las combinaciones cliente-producto que el modelo ya recomendó.  
   – Como `y_pred` se fija en 1 para todas esas filas, no existen falsos negativos y el recall resulta 1,0 por construcción.  
   – Este valor no refleja la cobertura real; es útil únicamente para verificar que, dentro del subconjunto recomendado, el código etiquetó correctamente los verdaderos positivos.

2. Recall “real” (0,84–0,91)  
   – Se calcula en `deep_batch_analysis`.  
   – Compara las recomendaciones con **todo** el universo de compras reales de la semana.  
   – Incluye los pares cliente-producto que ocurrieron pero no fueron recomendados; por ello incorpora falsos negativos y ofrece la medida de cobertura pertinente para el negocio.

Si bien se mostrarán ambas, debiesemos enfocarnos realmente en el Recall "real (0.84-0.91)" ya que:
- es la métrica de cobertura que responde a la pregunta central: “¿Cuántas compras efectivas logra captar el modelo?”.

**II. Escenario filtrado (3 200 primeras filas)**


````python
# --------- Construcción del DF filtrado (Top-N = 3200) --------------------
topN = 3200
def metrics_topN(pred, gt, N):
    m = evaluate_predictions(pred[:N], gt)            # accuracy, precision, recall, f1
    m['tasa_pred'] = N / len(gt)
    m['recall_real'] = recall_real(pred[:N], gt)
    return m

df_metrics_topN = pd.DataFrame([
    metrics_topN(pred1, combinations1, topN),
    metrics_topN(pred2, combinations2, topN),
    metrics_topN(pred3, combinations3, topN),
    metrics_topN(pred4, combinations4, topN)
], index=["1","2","3","4"])


# --------- Función auxiliar para dibujar líneas dobles --------------------
def plot_metric_compare(metric, ymin, ymax, ylabel):
    plt.figure(figsize=(6,3))
    sns.lineplot(x=df_metrics.index, y=df_metrics[metric],
                 marker="o", label=f"{metric} original")
    sns.lineplot(x=df_metrics_topN.index, y=df_metrics_topN[metric],
                 marker="o", linestyle="--", label=f"{metric} Top-N")
    plt.ylim(ymin, ymax)
    plt.xlabel("Semana")
    plt.ylabel(ylabel)
    plt.title(f"{ylabel} original vs Top-N")
    plt.grid(True)
    plt.legend()
    plt.show()

# --------- Gráficos solicitados -------------------------------------------
plot_metric_compare("precision", 0.0, 0.55, "Precisión")
plot_metric_compare("recall_real", 0, 1.1, "Recall real")
plot_metric_compare("f1", 0.0, 0.75, "F1")
plot_metric_compare("accuracy", 0.0, 0.55, "Accuracy")
````

3. Evolución temporal del modelo - Los resultados mostraron que el modelo mantuvo estabilidad en cobertura (0.84-0.91) pero con precisión baja, evidenciando la necesidad de políticas de filtrado más sofisticadas.

4. Impacto del reentrenamiento - El análisis comparativo reveló que los reentrenamientos semanales no mejoraron significativamente las métricas, sugiriendo que el modelo ya había convergido.

5. Valor del análisis comparativo - La comparación entre escenarios con y sin filtro Top-N demostró cómo un simple cambio de política puede multiplicar la precisión por 18x.

6. Valor del MLOps end-to-end  
   Construir un DAG reproducible con Airflow, Optuna y MLflow mostró cómo pasar de un notebook exploratorio a un servicio automatizado y auditable.

7. Gestión del drift y del hold-out  
   Un detector demasiado sensible (Como en nuestro caso) dispara reentrenamientos costosos; además, el conjunto de test debe mantenerse “puro” para reflejar el rendimiento real. La monitorización continua es tan importante como el modelo mismo.

8. Aprendizajes:

A lo largo del desarrollo, este proyecto permitió experimentar de manera práctica el ciclo completo de un sistema de ciencia de datos en un contexto realista, desde la construcción de pipelines de preprocesamiento y entrenamiento hasta la automatización y el monitoreo en producción. Se evidenció la importancia de balancear cobertura y precisión en sistemas de recomendación: maximizar la cobertura sin un control adecuado de la precisión puede traducirse en una sobrecarga operativa y pérdida de relevancia comercial. Además, quedó claro que la ingeniería de features, especialmente aquellas relacionadas con la recencia y frecuencia de compra, tiene un impacto decisivo en el desempeño de este escenario en particular, muchas veces por sobre la elección de algoritmos o el ajuste fino de hiperparámetros. Finalmente, la integración de herramientas de MLOps como Airflow y MLflow no solo facilitó la reproducibilidad y el seguimiento de experimentos, sino que también acercó la experiencia a los estándares de la industria, mostrando el valor de una gestión ordenada y colaborativa en proyectos de ciencia de datos.
