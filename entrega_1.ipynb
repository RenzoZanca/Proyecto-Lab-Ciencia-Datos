### Creaci√≥n de label

La idea sera agrupar los datos por cliente-producto-semana y para cada combinaci√≥n, determinar si hubo una compra (etiqueta positiva) o no (etiqueta negativa)

````python
# creacion de atributo semana
transactions['week'] = transactions['purchase_date'].dt.isocalendar().week
transactions.loc[transactions['purchase_date'] > '2024-12-29', 'week'] = 53 # caso borde
````

````python
import itertools

# Obtener los valores √∫nicos
product_ids = transactions['product_id'].unique()
customer_ids = transactions['customer_id'].unique()
weeks = transactions['week'].unique()

# Crear todas las combinaciones posibles
combinations = pd.DataFrame(list(itertools.product(product_ids, customer_ids, weeks)),
                            columns=['product_id', 'customer_id', 'week'])

# Unir con df_pre para obtener las columnas adicionales
df_combined = combinations.merge(transactions, on=['product_id', 'customer_id', 'week'], how='left')

# Agregar la etiqueta: 1 si existe en df_pre, 0 si no
df_combined['label'] = df_combined['order_id'].notna().astype(int)


df_combined = df_combined.sort_values(['customer_id','product_id','week'])

#  Suma de items del cliente‚Äêproducto en la semana anterior.
df_combined['items_last_week'] = (
    df_combined
      .groupby(['customer_id','product_id'])['items']
      .shift(1)
      .fillna(0)
)

# Promedio movil de items en las ultimas 4 semanas
df_combined['items_roll4_mean'] = (
    df_combined
      .groupby(['customer_id','product_id'])['items']
      .rolling(window=4, min_periods=1)
      .mean()
      .reset_index(level=[0,1], drop=True)
)
````

````python
df_combined['label'].value_counts()
````

Notamos que hay un desbalance muy grande de clases. Para solucionarlo hacemos subsampling de la clase 0 para balancear y llegar a 4 veces la cantidad de ejemplos positivos. Esto se realizar√° posteriormente, sobre los datos de entrenamiento.

### Cruce de informaci√≥n

````python
# 1) Cruzar informaci√≥n de clientes
tx = df_combined.merge(
    clients,
    on="customer_id",
    how="left",
    validate="many_to_one",  # cada transacci√≥n debe corresponder a un √∫nico cliente
    indicator="client_join"
)
print(tx['client_join'].value_counts())
#   Todas debiesen ser 'both'; si hay 'left_only', faltan clientes en el CSV


#2) Cruzar informaci√≥n de productos
tx = tx.merge(
    products,
    on="product_id",
    how="left",
    validate="many_to_one",  # cada transacci√≥n a un √∫nico producto
    indicator="product_join"
)
print(tx['product_join'].value_counts())
#   idealmente s√≥lo 'both'; si sale 'left_only', hay transacciones con product_id desconocido


#3) Quitamos las columnas join
tx = tx.drop(columns=["client_join", "product_join"])

print("--"*20)
print("Clientes en transacciones vs clientes √∫nicos:")
print(" ‚Üí transacciones:", tx['customer_id'].nunique())
print(" ‚Üí clientes.parquet:",  clients['customer_id'].nunique())

print("Productos en transacciones vs productos √∫nicos:")
print(" ‚Üí transacciones:", tx['product_id'].nunique())
print(" ‚Üí productos.parquet:",  products['product_id'].nunique())
print("--"*20)
````

````python
# 4) Filtrado de columnas

keep = [
    'purchase_date','customer_id','product_id','week',
    # Features hist√≥ricas
    'items_last_week','items_roll4_mean',
    # de clients:
    'customer_type','Y','X','num_deliver_per_week','custom_zone',
    # de products:
    'brand','category','sub_category','segment','package','size',
    # label
    'label'
]
df_final = tx[keep].copy()


# 5) Procesamiento de tipos
df_final['purchase_date'] = pd.to_datetime(df_final['purchase_date'])
df_final[['customer_type','brand','category','sub_category','segment','package']] = \
    df_final[['customer_type','brand','category','sub_category','segment','package']].astype('category')

# 6) Vista final
print("Dataframe pre-procesado:")
print(df_final.info())
df_final.head()
````

````python
# 3. Calidad de los datos

df_q = df_final.copy()
df_t = transactions.copy()
df_p = products.copy()
df_c = clients.copy()

def quality_report(df, name="DataFrame"):
    print(f"\n=== Reporte de calidad: {name} ===")
    # 1) Columnas y tipos
    print("\nColumnas:\n", df.dtypes, sep="")
    # 2) Valores nulos
    print("\n=== Valores faltantes por columna ===")
    print(df.isnull().sum())
    # 3) Duplicados
    n_dup = df.duplicated().sum()
    print(f"\n=== Filas duplicadas ===\n{n_dup} duplicados encontrados")
    # 4) Resumen num√©rico
    num_df = df.select_dtypes(include="number")
    if not num_df.empty:
        print("\n=== Estad√≠sticas resumen de columnas num√©ricas ===")
        print(num_df.describe().T)
        # 5) Conteo de outliers por IQR
        print("\n=== Conteo de outliers (IQR 1.5) por variable num√©rica ===")
        Q1 = num_df.quantile(0.25)
        Q3 = num_df.quantile(0.75)
        IQR = Q3 - Q1
        for col in num_df.columns:
            low, high = Q1[col] - 1.5 * IQR[col], Q3[col] + 1.5 * IQR[col]
            n_out = num_df[(num_df[col] < low) | (num_df[col] > high)].shape[0]
            print(f"  {col}: {n_out}")
    else:
        print("\n(No hay columnas num√©ricas en este DataFrame)")
    print("\n" + "-"*60)
    print(df.info())
````

## üìå Holdout [0.25 puntos]

Para evaluar correctamente el modelo y garantizar su capacidad de generalizaci√≥n, se deben dividir los datos en tres conjuntos:
- `Entrenamiento` : Para ajustar los par√°metros.
- `Validaci√≥n`: Para optimizar hiperpar√°metros y seleccionar el mejor modelo.
- `Prueba` : Para evaluar el rendimiento final en datos no vistos.

üëÄ **Hint**: *Recuerde que los datos tienen una temporalidad que debe considerarse al momento de separarlos, para evitar fugas de informaci√≥n. Es importante justificar la estrategia de partici√≥n elegida y visualizar la distribuci√≥n temporal de los conjuntos generados*

````python
#Checkeamos que la columna purchase_date sea de tipo datetime
df_final_ord = df_final.sort_values("week")

# Calculamos los puntos de corte en el tiempo (70% ‚Äì 15% ‚Äì 15%)

train_cut = df_final_ord["week"].quantile(0.70)
val_cut   = df_final_ord["week"].quantile(0.85)

print(train_cut)
print(val_cut)

train_df = df_final_ord[df_final_ord["week"] <= train_cut]
val_df   = df_final_ord[(df_final_ord["week"] > train_cut) & (df_final_ord["week"] <= val_cut)]
test_df  = df_final_ord[df_final_ord["week"] > val_cut]

# Revisamos tama√±os
print("Train:", len(train_df),
      "  Val:",   len(val_df),
      "  Test:",  len(test_df))
print("Train:", train_df['label'].value_counts())



counts = df_final_ord.groupby("week").size()
plt.figure(figsize=(8,3))
counts.plot(marker="o")
plt.axvline(train_cut, color="C1", linestyle="--", label="fin train")
plt.axvline(val_cut,   color="C2", linestyle="--", label="fin val")
plt.legend()
plt.title("Conteo semanal de transacciones y cortes temporales")
plt.show()
````

````python
from sklearn.utils import resample

def temporal_undersample(df, ratio=4, time_col='week', label_col='label', random_state=42):
    """Aplica undersampling estratificado por semana, manteniendo todos los positivos y una proporci√≥n limitada de negativos."""
    parts = []
    for week, grp in df.groupby(time_col):
        pos = grp[grp[label_col] == 1]
        neg = grp[grp[label_col] == 0]
        n_neg = min(len(neg), len(pos) * ratio)
        neg_sub = resample(neg, replace=False, n_samples=n_neg, random_state=random_state)
        parts.append(pd.concat([pos, neg_sub]))
    return pd.concat(parts).sample(frac=1, random_state=random_state)

# Aplicar undersampling al set de entrenamiento
train_df = temporal_undersample(train_df, ratio=4)
# Revisamos tama√±os
print("Train:", len(train_df),
      "  Val:",   len(val_df),
      "  Test:",  len(test_df))
print("Train:", train_df['label'].value_counts())
````

#### Evitamos look-ahead bias
En un problema de demanda semanal, si mezcl√°ramos al azar registros en el entrenamiento, el modelo ver√≠a informaci√≥n del futuro (por ejemplo, promociones de fin de a√±o, antes.) Al usar cortes temporales, evitamos esto y se "simula la realidad" : entreno en lo m√°s antiguo, valido en lo intermedio y pruebo en lo m√°s reciente.

Cada una de estas transformaciones debe ser justificada en funci√≥n de su relevancia para el problema y los datos, y es importante evaluar su impacto en el rendimiento del modelo. Adem√°s, el pipeline debe ser flexible y modular para poder probar diferentes configuraciones de preprocesamiento.

Manejo de fechas:

````python
from datetime import datetime, timedelta
import random

# Funci√≥n que recibe una semana y devuelve la fecha de algun dia al azar de esa semana
def semana_a_fecha_random(week, year=2024):
    # Clip entre 1 y 52 para evitar week 53 inv√°lida
    w = min(max(week, 1), 52)
    monday = datetime.fromisocalendar(year, w, 1)
    return monday + timedelta(days=random.randint(0, 6))

def extract_date_features(X):
    X = X.copy()
    def pick_date(row):
        if pd.notna(row["purchase_date"]):
            return row["purchase_date"]
        return semana_a_fecha_random(int(row["week"]))
    X["date"]  = X.apply(pick_date, axis=1)
    X["month"] = X["date"].dt.month.astype("category")
    X["day"]   = X["date"].dt.day.astype("category")
    return X.drop(columns="purchase_date")

date_transformer = FunctionTransformer(extract_date_features)
````

### Justificaci√≥n de las transformaciones:

1. **Extracci√≥n de caracter√≠sticas**

  * Capturar la cantidad de transacciones por cliente para que el modelo pueda interpretar que tanto compra un cliente.
   * Para cada cliente computar cuantos productos diferentes y en que cantidades compra.
   * Para cada producto, calcular cada cuanto tiempo se compra, pues puede influir en que tan probable es comprarlo.
   


2. **Extracci√≥n de fecha (`purchase_date` ‚Üí `month`, `day`)**

   * Captura patrones estacionales (picos por mes o d√≠a) sin tratar la fecha completa como un n√∫mero ordinal.
   * Convertir a `category` evita que el modelo interprete ‚Äú12 > 1‚Äù de forma err√≥nea; cada valor es un grupo independiente.
   * Rellenar las fechas faltantes con alg√∫n dia de esa semana.
  

3. **Imputaci√≥n num√©rica con la mediana**

   * Rellena faltantes de forma robusta ante outliers (p. ej. valores extremos de `items` o coordenadas).
   * Evita que valores extremos sesguen el ‚Äúvalor t√≠pico‚Äù que usar√≠a la media.

4. **Escalado Min‚ÄìMax de variables num√©ricas**

   * Reescala todas las variables a \[0, 1], unificando su rango.
   * Conserva la forma de la distribuci√≥n y no ‚Äúaplana‚Äù tanto los outliers.

5. **Imputaci√≥n categ√≥rica con la moda**

   * Asigna a los faltantes la categor√≠a m√°s frecuente, opci√≥n conservadora que no introduce nuevas etiquetas de baja frecuencia.
   * Reduce el riesgo de overfitting.

6. **One-Hot Encoding para variables categ√≥ricas**

   * Evita imponer un orden num√©rico entre niveles (una codificaci√≥n ordinal distorsionar√≠a relaciones).
   * Genera variables binarias f√°ciles de interpretar y adecuadas para cualquier modelo.

7. **ColumnTransformer + Pipeline**

   * Agrupa y a√≠sla la l√≥gica de cada tipo de variable en un √∫nico objeto, mejorando legibilidad y evitando fugas de datos.
   * Permite encadenar todo el flujo (fecha ‚Üí imputaci√≥n ‚Üí encoding ‚Üí escalado) y reutilizarlo en `fit`/`predict` o en b√∫squeda de hiperpar√°metros sin reescribir c√≥digo.

````python
import optuna
from optuna.samplers import TPESampler
from optuna.pruners import MedianPruner
from optuna.integration import XGBoostPruningCallback
import xgboost as xgb
from sklearn.metrics import f1_score
import numpy as np
from sklearn.metrics import classification_report

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)


# 1) Preparar DMatrix
dtrain = xgb.DMatrix(X_train_tr, label=y_train)
dval   = xgb.DMatrix(X_val_tr,   label=y_val)

# 2) Definir funci√≥n objetivo para Optuna
def objective(trial):
    param = {
        "learning_rate":    trial.suggest_loguniform("learning_rate",    0.10, 0.25),
        "max_depth":        trial.suggest_int("max_depth",               8,   16),
        "subsample":        trial.suggest_float("subsample",             0.65, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree",      0.6,  1.0),
        "gamma":            trial.suggest_float("gamma",                 0.0,  2.0),
        "min_child_weight": trial.suggest_int("min_child_weight",        1,   10),
        "reg_alpha":        trial.suggest_loguniform("reg_alpha",       0.1,   5.0),
        "reg_lambda":       trial.suggest_loguniform("reg_lambda",      0.1,   5.0),
        "tree_method":      "hist",
        "verbosity":        0,
        "eval_metric":      "logloss",
        "objective":        "binary:logistic"
    }
    n_rounds = trial.suggest_int("n_estimators", 50, 300)

    pruning_callback = XGBoostPruningCallback(trial, "validation-logloss")

    booster = xgb.train(
        params=param,
        dtrain=dtrain,
        num_boost_round=n_rounds,
        evals=[(dval, "validation")],
        callbacks=[
            pruning_callback,
            xgb.callback.EarlyStopping(rounds=30)
        ],
        verbose_eval=False,
    )

    preds = (booster.predict(dval) > 0.5).astype(int)
    return f1_score(y_val, preds, pos_label=1)

# 3) Ejecutar estudio Optuna
study = optuna.create_study(
    direction="maximize",
    sampler=TPESampler(seed=42),
    pruner=MedianPruner(n_startup_trials=5)
)
study.optimize(objective, timeout=300)

# 4) Mostrar resultados
best = study.best_trial
print(f"‚Üí Trials completados: {len(study.trials)}")
print(f"‚Üí Mejor F1 (clase 1) en validaci√≥n (threshold 0.5): {best.value:.3f}")
print("‚Üí Hiperpar√°metros √≥ptimos:")
for k, v in best.params.items():
    print(f"   ‚Ä¢ {k}: {v}")

# 5) Re-entrenar booster final con los mejores hiperpar√°metros
params_final = best.params.copy()
params_final.update({
    "objective": "binary:logistic",
    "eval_metric": "logloss",
    "verbosity": 0,
    "tree_method": "hist"
})
n_rounds_final = best.params["n_estimators"]

booster_final = xgb.train(
    params=params_final,
    dtrain=dtrain,
    num_boost_round=n_rounds_final,
    evals=[(dval, "validation")],
    early_stopping_rounds=30,
    verbose_eval=False
)

# 6) B√∫squeda de umbral √≥ptimo en validaci√≥n
probs_val = booster_final.predict(dval)
best_thr, best_f1 = 0.0, 0.0
for thr in np.linspace(0.1, 0.9, 81):
    preds_thr = (probs_val > thr).astype(int)
    f1 = f1_score(y_val, preds_thr, pos_label=1)
    if f1 > best_f1:
        best_f1, best_thr = f1, thr

print(f"‚Üí Mejor F1 ajustado en validaci√≥n: {best_f1:.3f} con umbral = {best_thr:.2f}")

# 7) Evaluaci√≥n final en test usando ese umbral
dtest = xgb.DMatrix(X_test_tr)
probs_test = booster_final.predict(dtest)
y_pred_test = (probs_test > best_thr).astype(int)

print("\nReporte en TEST:")
print(classification_report(y_test, y_pred_test))
````

#### Predicciones particulares

````python
# Predicciones particulares:

from alibi.explainers import AnchorTabular

features = final_pipeline.named_steps['features']
X_train_processed = features.transform(X_train)
X_test_processed = features.transform(X_test)
model = final_pipeline.named_steps['classifier'].fit(X_train_processed, y_train)
predict_fn = lambda x: model.predict(x) # creamos funci√≥n de predicci√≥n
explainer = AnchorTabular(predict_fn, model.feature_names_in_, seed=42) # instanciamos Anchor usando funci√≥n creada
explainer.fit(X_train_processed.to_numpy())

import operator

def rules_system(idx):
    ops = {
        '=': operator.eq,
        '<=': operator.le,
        '>=': operator.ge,
        '<': operator.lt,
        '>': operator.gt
    }

    obs = np.array(X_test_processed.iloc[idx], ndmin=2)
    explanation = explainer.explain(obs, threshold=0.95)
    anchor = explanation.anchor
    pred_label = explainer.predictor(obs)[0]

    print(f"--- Observaci√≥n {idx} ---")
    print('Prediction: ', pred_label)
    print('Anchor: %s' % (' AND '.join(anchor)))

    covered = X_test_processed.copy()
    for feature in anchor:
        matched = False
        for symbol, func in ops.items():
            if f' {symbol} ' in feature:
                feat, val = feature.split(f' {symbol} ')
                val = type(X_test_processed[feat].iloc[0])(val)
                covered = covered[func(covered[feat], val)]
                matched = True
                break
        if not matched:
            print(f"[!] Regla ignorada: {feature}")

    total_covered = len(covered)
    correct_preds = np.sum(model.predict(covered) == pred_label)
    global_precision = correct_preds / total_covered if total_covered > 0 else 0

    print(f"Poblaci√≥n explicada por la regla: {total_covered} ejemplos")
    print(f"Precisi√≥n en poblaci√≥n cubierta: {global_precision:.2f}")
    print()


idxs = [1000, 3001, 5751]
for idx in idxs:
    rules_system(idx)
````
